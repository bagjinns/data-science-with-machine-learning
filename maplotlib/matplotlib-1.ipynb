```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'b-')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'r')
```

```python
# OO
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.1, 0.4, 0.4])

axes1.plot(x, y)
axes1.set_title("Big plot")
axes2.plot(y, x)
axes2.set_title("small plt")
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
axes[0].plot(x, y)
axes[0].set_title("First graph")

axes[1].plot(y, x)
axes[1].set_title("Second graph")
```

```python
# fig = plt.figure(figsize=(3, 2))
# ax = fig.add_axes([0,0, 1, 1])
# ax.plot(x, y)
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 3))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
fig.savefig("My_fig.png")
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_axes([0,0, 1, 1])
ax.plot(x, x**2, label="X^2")
ax.plot(x, x**3, label="X^3")

ax.legend()
plt.tight_layout()
```
